```python
import pandas as pd

# 載入 Excel 檔案
excel_file_path = 'C:/Users/wei/Documents/實習/RebasCPBL/中職2023年-逐球數據-20230906.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name3 = 'AWAY-打席資訊'  # 請替換成你要載入的工作表名稱
sheet_name4 = 'HOME-打席資訊'
```

```python
baseball_cols = ['比賽編號','局數','壘包狀態','客場','主場','打席結果','打者','RE','WPA','RE24'] 
```

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name3)
away
```

```python
away_play = away[baseball_cols]
away_play['局數'] = away_play['局數'].replace({1: '1上', 2: '2上', 3: '3上', 4: '4上', 5: '5上', 6: '6上',7:'7上',8:'8上',9:'9上',10:'10上',11:'11上',12:'12上'})
away_play
```

```python
# 将"客場"改名为"球隊"
away_play.rename(columns={'客場': '球隊'}, inplace=True)

# 删除"主場"列
away_play.drop(columns=['主場'], inplace=True)
away_play
```

```python
Home = pd.read_excel(excel_file_path, sheet_name=sheet_name4)
Home
```

```python
Home_play = Home[baseball_cols]
Home_play['局數'] = Home_play['局數'].replace({1: '1上', 2: '2上', 3: '3上', 4: '4上', 5: '5上', 6: '6上',7:'7上',8:'8上',9:'9上',10:'10上',11:'11上',12:'12上'})
Home_play
```

```python
Home_play.rename(columns={'主場': '球隊'}, inplace=True)

# 删除"主場"列
Home_play.drop(columns=['客場'], inplace=True)
Home_play
```

```python
merged_df = pd.concat([Home_play, away_play], ignore_index=True)

# 輸出合併後的資料表
merged_df
```

```python
included_pitchers = ['SH','SH_FC','SH_E']

# Filter the DataFrame
filtered_df = merged_df[merged_df['打席結果'].isin(included_pitchers)]
```

```python
filtered_df
```

```python
agg_functions = {'WPA': 'sum', 'RE':'sum' ,'打席結果':'count'}

result = filtered_df.groupby('打者').agg(agg_functions).reset_index()
result.rename(columns={'打席結果': '犧牲短打次數','RE': '犧牲短打累積出的RE24'}, inplace=True)
result= result.sort_values(by=['WPA', '犧牲短打次數'], ascending=[True, False])
result
```

```python
#result.to_excel("C:/Users/wei/Documents/實習/中職短打/player_result.xlsx")
```

```python
result2 = filtered_df.groupby('球隊').agg(agg_functions)
result2.rename(columns={'打席結果': '犧牲短打次數','RE': '犧牲短打累積出的RE24'}, inplace=True)
result2= result2.sort_values(by=['WPA', '犧牲短打次數'], ascending=[True, False])
result2
```

```python
#result2.to_excel("C:/Users/wei/Documents/實習/中職短打/Team_result.xlsx")
```

```python
result3 = filtered_df.groupby('局數').agg(agg_functions).reset_index()
result3.rename(columns={'打席結果': '犧牲短打次數','RE': '犧牲短打累積出的RE24'}, inplace=True)
result3= result3.sort_values(by=['局數', 'WPA'], ascending=[True, False])
result3
```

```python
result4 = filtered_df.groupby(['局數','球隊','壘包狀態']).agg(agg_functions).reset_index()
result4.rename(columns={'打席結果': '犧牲短打次數','RE': '犧牲短打累積出的RE24'}, inplace=True)
result4= result4.sort_values(by=['局數', 'WPA'], ascending=[True, False])
result4
```
